## Clean the Data

```python
df.rename(columns = {'NAME' : 'name'}, inplace = True)
```

```python
# Dropping unnessesary columns


df = df.drop(['id','host id','house_rules','license'], axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
# Check for missing values

missing_values = df.isnull().sum()
```

```python
missing_values
```

```python
# List of column names to check for NaN values

cols_to_check = ['name', 'host_identity_verified', 'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country', 'country code', 'instant_bookable', 'cancellation_policy', 'Construction year', 'price', 'service fee', 'minimum nights', 'number of reviews', 'last review', 'reviews per month', 'review rate number', 'calculated host listings count', 'availability 365']
```

```python
cols_to_check
```

```python
df_clean = df[df[cols_to_check].notnull().all(axis=1)]
```

```python
df_clean.shape
```

```python
# Find duplicates

df_dups = df_clean[df_clean.duplicated()]
```

```python
df_dups.shape
```

```python
df = df_clean.drop_duplicates()
```

```python
df.shape
```

```python
#Check mixed-type columns 

for col in df.columns.tolist():
      weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
        
# No mixed-type columns
```
